```python
import pandas as pd

df = pd.read_csv('../data/products.csv')
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print(df.head(3))
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.to_csv('../data/products.csv', index=False)
print("Products cleaned. Rows remaining:", len(df))
```

```python
import json

# Read your chatbot.json file
with open('../data/chatbot.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Print the structure so we can see what's inside
print("Type:", type(data))
print("Keys (if dict):", data.keys() if isinstance(data, dict) else "It's a list")

# Show first item
if isinstance(data, list):
    print("First item:", data[0])
elif isinstance(data, dict):
    first_key = list(data.keys())[0]
    print("First key:", first_key)
    print("First value:", data[first_key][0] if isinstance(data[first_key], list) else data[first_key])
```

```python
df_reviews = pd.read_csv('../data/reviews.csv')

print("Reviews Columns:", df_reviews.columns.tolist())
print(df_reviews.head(3))

# Use 'Text' instead of 'reviewText'
df_reviews.dropna(subset=['Text'], inplace=True)
df_reviews.drop_duplicates(inplace=True)

df_reviews['Text'] = df_reviews['Text'].str.lower()
df_reviews['Text'] = df_reviews['Text'].str.replace(
    '[^a-zA-Z\s]', '', regex=True)

# Reduce size (faster processing)
df_reviews = df_reviews.head(2000)

df_reviews.to_csv('../data/reviews.csv', index=False)

print("Reviews cleaned. Rows:", len(df_reviews))
```
